# color Detection

```python
def color_detect(img):
    img2=img.copy()
    m,n,q = img.shape
    for i in range(m):
        for j in range(n):
            if img[i,j,0]<70 and img[i,j,1]<70 and img[i,j,2]<70:
                img2[i,j,0]=255
                img2[i,j,1]=255
                img2[i,j,2]=255
    return img2
```
